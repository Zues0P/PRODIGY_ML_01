Plotting the heatmap for better understanding of the missing values 

```python
sns.heatmap(test_data.isnull(),yticklabels=False,cbar=False)
```

```python
#filling in missing values
numeric_columns = test_data.select_dtypes(include=['int64', 'float64']).columns.tolist()
# grouping all the numeric columns int and float both 
test_data[numeric_columns] = test_data[numeric_columns].fillna(test_data[numeric_columns].mean())
#using fillna to fill the missing values using mean values from the column

#similarly grouping all the object64 type columns 
object_columns = test_data.select_dtypes(include='object').columns.tolist()

test_data[object_columns] = test_data[object_columns].apply(lambda x: x.fillna(x.mode()[0]))

#after filling the missing values printing the result to check if filled or not
missing_values_after_filling = test_data.isnull().sum()
print(missing_values_after_filling)
```

```python
sns.heatmap(test_data.isnull(),yticklabels=False,cbar=False)
```

all missing values filled therefore moving forward and dropping all the columns which are not needed as they had multiple missing values before the fillna apply cation  so dropping them  

```python
test_data.drop(['Alley','PoolQC','Fence','MiscFeature','GarageYrBlt'],axis=1,inplace=True)
```
